```python
category_mapping = {'Fiction' : "Fiction",
 'Juvenile Fiction': "Children's Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Children's Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction"}

books["simple_categories"] = books["categories"].map(category_mapping)
```

```python
from transformers import pipeline

candidate_labels = ["Fiction","Nonfiction"]

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli",device=0)
```

#### First testing for known label

```python
sequence_to_classify=books.loc[books["simple_categories"]=="Fiction","description"].reset_index(drop=True)[0]
sequence_to_classify
```

```python
classifier(sequence_to_classify, candidate_labels)
```

```python
import numpy as np

max_label_index=np.argmax(classifier(sequence_to_classify, candidate_labels)['scores'])
max_label=classifier(sequence_to_classify, candidate_labels)["labels"][max_label_index]
max_label
```

```python
def classifier_function(sequences,candidate_labels):
    predictions=classifier(sequences, candidate_labels)
    max_label_index=np.argmax(predictions["scores"])
    max_label=predictions["labels"][max_label_index]
    return max_label
```

```python
from tqdm import tqdm

actual_f=[]
predicted_f=[]

for i in tqdm(range(0,300)):
    sequence_to_classify=books.loc[books["simple_categories"]=="Fiction","description"].reset_index(drop=True)[i]
    predicted_f+=[classifier_function(sequence_to_classify,candidate_labels)]
    actual_f+=['Fiction']
```

```python
actual_nf = []
predicted_nf = []

for i in tqdm(range(0, 300)):
    sequence_to_classify = books.loc[
        books["simple_categories"] == "Nonfiction", "description"
    ].reset_index(drop=True)[i]
    predicted_nf += [classifier_function(sequence_to_classify, candidate_labels)]
    actual_nf += ["Nonfiction"]
```

```python
predictions_df = pd.DataFrame({"actual_categories": actual_f+actual_nf, "predicted_categories": predicted_f+predicted_nf})
predictions_df
```

```python
predictions_df["correct_prediction"] = np.where(
    predictions_df["actual_categories"] == predictions_df["predicted_categories"], 1, 0
)
predictions_df["correct_prediction"].sum()/len(predictions_df)*100
```

```python
nonclassified_categories = books.loc[books['simple_categories'].isna(),["isbn13","description"]].reset_index(drop=True)
nonclassified_categories.head(),nonclassified_categories.shape
```

```python
predicted_category=[]
isbns=[]

for i in tqdm(range(0, len(nonclassified_categories))):
    sequence_to_classify = nonclassified_categories["description"][i]
    predicted_category+= [classifier_function(sequence_to_classify, candidate_labels)]
    isbns+=[nonclassified_categories['isbn13'][i]]
```

```python
predicted_nonclassified_categories = pd.DataFrame(
    {"isbn13": isbns, "predicted_categories": predicted_category}
)
```

```python
predicted_nonclassified_categories
```

```python
books=pd.merge(books,predicted_nonclassified_categories,on="isbn13",how="left")
books["simple_categories"]=np.where(books['simple_categories'].isna(),books['predicted_categories'], books['simple_categories'])
books.drop(['predicted_categories'],axis = 1, inplace = True)
```
