#### The model perfectly catches the emotion of the sentences.

```python
import numpy as np

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotions_label_scores={ label: [] for label in emotion_labels}

def max_emo_scores(preds):
    per_em_sc={label:[] for label in emotion_labels}
    for pred in preds:
        sor_pred=sorted(pred,key=lambda x : x["label"])
        for i , l  in enumerate(emotion_labels):
            per_em_sc[l].append(sor_pred[i]["score"])
    return { label: np.max(scores) for label , scores in per_em_sc.items()}
```

```python
for i in range(10):
    isbn.append(books["isbn13"][i])
    sentences=books["description"][i].split(".")
    preds= classifier(sentences)
    max_scores= max_emo_scores(preds)
    for label in emotions_label_scores:
        emotions_label_scores[label].append(max_scores[label])
```

```python
from tqdm import tqdm
emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]

isbn = []
emotion_scores = {label: [] for label in emotion_labels}


for i in tqdm(range(len(books))):
    isbn.append(books["isbn13"][i])
    sentences=books["description"][i].split(".")
    preds= classifier(sentences)
    max_scores= max_emo_scores(preds)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
emotions=pd.DataFrame(emotion_scores)
emotions["isbn13"]=isbn
```

```python
emotions.head()
```

```python
books=pd.merge(left = books, right = emotions , on="isbn13")
books.head()
```
